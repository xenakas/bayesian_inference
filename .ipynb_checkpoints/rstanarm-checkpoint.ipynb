#  Bayesian Regression Modeling

## Bayesian Linear Models

```r
songs = read.csv(file='https://assets.datacamp.com/production/repositories/2199/datasets/3c921f85674c92085b3428c303b9364573a8bd4f/datacamp-spotify-data.csv', header=TRUE, sep=",")
head(songs)
```

```r
lm_model <- lm(popularity ~ song_age, data = songs)
```

```r
summary(lm_model)
```

```r
library(broom)

tidy(lm_model)
```

```r
library(rstanarm)
```

```r
stan_model <- stan_glm( popularity ~ song_age, data = songs)
```

- sigma: Standard deviation of errors
- mean_PPD: mean of posterior predictive samples
- log-posterior: analogous to a likelihood
- Rhat: a measure of within chain variance compared to across chain variance (if all Rhat values are below 1.1, your model converged)

Frequentist: parameters are fixed, data is random
Bayesian: parameters are random, data is fixed

- Confidence interval: Probability that a range contains the true value
- Credible interval: Probability that the true value is within a range

```r
# Create the 90% credible intervals
posterior_interval(stan_model)

# Create the 95% credible intervals
posterior_interval(stan_model, prob = 0.95)

# Create the 80% credible intervals
posterior_interval(stan_model, prob = 0.8)
```

```r
confint(lm_model, parm = "song_age", level = 0.95)
```

```r
posterior_interval(stan_model, pars = "song_age", prob = 0.95)
```

```r
library(tidybayes)
posterior <- spread_draws(stan_model, song_age) 
head(posterior)
```

```r
library(dplyr)
```

```r
mean(between(posterior$song_age, -0.006, -0.005))
```

## Modifying a Bayesian Model

Posterior Distributions
- Posterior distributions sampled in groups called chains
- Each sample in a chain is an iteration

- Fewer iterations = shorter estimation time
- Not enough iteration = convergence problems

```r
summary(stan_model)
```

```r
# 3 chains, 1000 iterations, 500 warmup
model_3chains <- stan_glm(popularity ~ song_age, data = songs,
    chains = 3, iter = 1000, warmup = 500)

# Print a summary of model_3chains
summary(model_3chains)
```

```r
# 2 chains, 100 iterations, 50 warmup
model_2chains <- stan_glm(popularity ~ song_age, data = songs,
   chains = 2, iter = 100, warmup = 50)

# Print a summary of model_2chains
summary(model_2chains)
```

###### Estimation errors are threats to the validity of the model

> There were 15 divergent transitions after warmup. Increasing adapt_delta above 0.8 may help.

- Too big of steps in the estimator
- Adjust step size


```r
stan_model_adj <- stan_glm(popularity ~ song_age, data = songs,
  control = list(adapt_delta = 0.99)) # 0.95 by default (between 0 and 1), an increase delta decrease step size and resolve divergent transition error  
```

```r
summary(stan_model_flat)
```

> Chain 1 reached the maximum tree depth

- Sample evaluates branches and looks for a good place to "U-Turn" (in a series of possible branches)
- Max tree depth indicates poor efficiency (if the sampler reaches max tree depth before finding a good place to "U-turn", then the sample terminated the iteration before finding a good stopping place; therefore the posterior distribiution is not being sampled as efficiently as it should be)


```r
stan_model_tree <- stan_glm(popularity ~ song_age, data = songs,
  control = list(max_treedepth = 15))  # 10 by default, by increaing we allow the algorithm to look further for a good place to u-turn
```

```r
summary(stan_model_tree)
```

We have a total of 4000 iterations that make up our posterior distribution 

```r
predictions <- posterior_linpred(stan_model)
dim(predictions)
```

A matrix that has a row for each iteration and a column for each observation. Each new set of predicted scores is calculated based on a revious iteration.

Because we have 4000 iterations, we can calculate 4000 predicted Y's for each observation.

Because these values are generated using the model parameters, these predicted scores form a distribution of what our observed scores should look like _if the specified model were correct_

Thus deviations from these predictive distributions are an indicator of poor fit.

```r
head(predictions)
```

To look fom deviations compare the distribution of predicted scores in an iteration to the distribution of observed Y's. 

```r
# First replication
iter1 <- predictions[1,]
summary(iter1)
```

```r
# Second replication
iter2 <- predictions[2,]
summary(iter2)
```

- elpd_loo - loo estimate; 
- p_loo - the effective number of parameters in the model;
- looic - loo estimate converted to a deviane scale ("-2")

```r
summary(stan_model_alternative)
```

```r
loo_2pred = loo(stan_model_alternative)
loo_2pred
```

- positive score means 2nd model is favoured
- negative score means 1st model is favoured
- se helps to decide if the difference is meaningful (if abs of diff is less than se, models  perform   the same, choose simpler model)

## Visualizing a Bayesian Model

```r
tidy_coef <- tidy(stan_model)

model_intercept <- tidy_coef$estimate[1]
model_intercept

model_slope <- tidy_coef$estimate[2]
model_slope
```

```r
ggplot(songs, aes(x = song_age, y = popularity )) + 
  geom_point() +
  geom_abline(intercept = model_intercept, slope = model_slope)
```

```r
draws <- spread_draws(stan_model, `(Intercept)`, song_age)

draws
```

```r
ggplot(songs, aes(x = song_age, y = popularity )) + 
  geom_point() +
  geom_abline(data = draws, aes(intercept = `(Intercept)`, slope = song_age),  size = 0.2, alpha = 0.1, color = "skyblue") +
  geom_abline(intercept = model_intercept, slope = model_slope) 
```
